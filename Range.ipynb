```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
ev=pd.read_excel("range.xlsx")
ev
```

```python

ev=ev.drop(columns=['Unnamed: 13','Note','Fan','Unnamed: 8'],axis=True)
ev.head(3)
```

```python
ev['batterysoc']=ev['Battery State of Charge (Start)']-ev['Battery State of Charge (End)']
```

```python
ev
```

```python
ev=ev.drop(columns=['Battery State of Charge (Start)','Battery State of Charge (End)'],axis=True)
ev
```

```python
ev=ev.drop(columns=['Date'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ev['Trip']=ev['Trip'].astype(str)
ev['Trip']=le.fit_transform(ev['Trip'])
ev['Route/Area']=ev['Route/Area'].astype(str)
ev['Route/Area']=le.fit_transform(ev['Route/Area'])
ev['Weather']=ev['Weather'].astype(str)
ev['Weather']=le.fit_transform(ev['Weather'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_features = ev.select_dtypes(include=['object']).columns
print(categorical_features)
# Compute the correlation matrix
corr = ev.corr()

# Plot the heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
corr["Distance [km]"]
```

```python
X = ev[['batterysoc','Weather','Duration [min]','Route/Area' ]]
y =ev['Distance [km]']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
stand = scaler.fit_transform(X)
print(stand)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y, test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(mae)
print(mse)
print(f"R-squared (R2): {r2:.2f}")
```

```python
from sklearn.linear_model import Ridge
a=Ridge()
a.fit(x_train,y_train)
```

```python

y_pred=a.predict(x_test)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(mae)
print(mse)
print(f"R-squared (R2): {r2:.2f}")
```
